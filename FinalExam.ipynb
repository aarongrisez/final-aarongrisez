```python
import rossler as r
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numba as nb
from IPython.display import Image
plt.ioff()
```

This assignment was to evaluate the behavior of the Rossler Attractor, defined with the three coupled Ordinary Differential Equations:

$$ \dot{x} = -y -z$$
$$ \dot{y} = x + ay$$
$$ \dot{z} = b + z(x - c)$$

This investigation was to only vary the $c$ parameter, though very interesting results can come from varying $a$ and $b$ as well as the initial conditions.

```python
@nb.jit
def plot_many_c(n, c0, cf):
    delta_c = (cf - c0) / n
    plt.figure(1, figsize=(12,8))
    plt.xlabel('$c$')
    plt.xlim((2,6))
    plt.ylabel('$x$ Positions ($m$)')
    plt.ylim((3,12))
    plt.title('$x$ Relative Maxima for varying $c$')
    for i in range(n):
        c_rossler(c0, cf, delta_c, i)
    plt.show()
    
@nb.jit
def c_rossler(c0, cf, delta_c, i):
    c_current = c0 + i * delta_c
    sol = r.rossler(c_current, dt=0.02, T=300)
    mx = r.find_maxima('x', sol, N=150)
    cs = np.zeros(mx.size) + c_current
    plt.scatter(cs, mx, marker='.', s=0.1)
```

What is particularly interesting about this chaotic system as opposed to the logistic map, for example, is its time dependence. Rather than being a discrete update map with one dimension, this is a continuous parametric function in three dimensions.

The first interesting feature is that each orbit investigated is bounded. They may not be periodic, but their motion is limited to a finite volume for the entirety of the evolution (for the time interval investigated--certainly this is not an analytic proof that the orbits will continue to be bounded, but such a proof is outside the scope of this investigation.)

Qualitatively, we see periodic orbits for some values of c, while others begin to display aperiodicity.
