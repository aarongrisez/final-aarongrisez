```python
import rossler as r
import pandas as pd
import matplotlib.pyplot as plt
plt.ioff()
```

```python
sol2 = pd.read_csv('data/c_equals_2')
#sol3 = pd.read_csv('data/c_equals_3')
#sol4 = pd.read_csv('data/c_equals_4')
sol4_15 = pd.read_csv('data/c_equals_4_15')
#sol4_25 = pd.read_csv('data/c_equals_4_25')
sol5_7 = pd.read_csv('data/c_equals_5_7')
```

This assignment was to evaluate the behavior of the Rossler Attractor, defined with the three coupled Ordinary Differential Equations:

$$ \dot{x} = -y -z$$
$$ \dot{y} = x + ay$$
$$ \dot{z} = b + z(x - c)$$

This investigation was to only vary the $c$ parameter, though very interesting results can come from varying $a$ and $b$ as well as the initial conditions. First, we plot a single dimension as a time plot for $c = 2$.

### $C = 2$

```python
r.plotxyz_separate(sol2)
plt.show()
```

This set of plots shows that the orbits can be projected onto each dimension and behave predominantly like oscillators. However, changing the $c$ value to $4.15$ for example, we begin to see more complicated behavior.

```python
r.plotxyz_separate(sol4_15)
plt.show()
```

These changing oscillations resemble the same sort of beginnings to chaotic behavior that we saw with the logistic map. The behavior is still mostly periodic, but in larger cycles than before. Particularly, we can see some beginnings of what may be bifurcation of vibrational modes if we look at the relative maxima in each dimension.

Below is a set of parametric plots for the orbits when the initial conditions are $x=y=z=0$ and $c=2$.

```python
r.plot_2D_parametric(sol2)
plt.show()
```

What is particularly interesting about this chaotic system as opposed to the logistic map, for example, is its dimension. Rather than being a discrete update map with one dimension, this is a continuous parametric function in three dimensions. These parametric plots begin to show how each dimension is related to the rest.

An interesting feature is that each orbit investigated is bounded. They may not be periodic, but their motion is limited to a finite volume for the entirety of the evolution (for the time interval investigated--certainly this is not an analytic proof that the orbits will continue to be bounded, but such a proof is outside the scope of this investigation.)

Qualitatively, we see periodic orbits for some values of c, while others begin to display aperiodicity. For example, in the following two sets of plots, the $c=4.15$ plot shows what appears to be 3 pairs of stable orbits. However, in the $c=5.7$ plots, we see no particularly stable orbits.

### $C = 4.15$

```python
r.plot_2D_parametric(sol4_15)
plt.show()
```

### $C = 5.7$

```python
r.plot_2D_parametric(sol5_7)
plt.show()
```

Now, looking at a 3-dimensional parametric plot of the solutions, we see still more confirmation for periodic, non-chaotic orbits for low values of $c$ with aperiodicity appearing for larger values of $c$.

As with the logistic map, an asymptotic analysis seems to appear useful for comparing behavior with different $c$ values. Here, we look at the relative maxima in the $x$ dimension ignoring transient behavior. Similar structure plots are likely to occur in the $y$ and $z$ dimensions. The dark black line suggests the convergence of the asymptotic orbit to a stable, periodic one. However, just below $c = 3$, we see a bifurcation in the relative maxima. Roughly half of the time, the orbit will be in the $x = 5$ relative maximum, and the other half of the time it will be in the $x = 6$ maximum. It is this cascading effect we observe as $c$ increases that is so characteristic of chaotic systems. As with the logistic map, there seem to be points of a sort of "resonance". In this case, one would appear to be at approximately $c = 5.3$. I still have yet to understand why these points of clarity occur in the middle of chaos, but that is a matter for futher exploration.
